```python
from PIL import ImageEnhance
```

```python
plt.figure(figsize=(10,10))
image_color_enhance = image.copy()
image1 = ImageEnhance.Color(image_color_enhance).enhance(.5)
image2 = ImageEnhance.Contrast(image_color_enhance).enhance(.5)
image3 = ImageEnhance.Brightness(image_color_enhance).enhance(15)
image4 = ImageEnhance.Sharpness(image_color_enhance).enhance(.5)



plt.subplot(2,2,1)
plt.imshow(image4)
plt.title("Sharpness")



```
