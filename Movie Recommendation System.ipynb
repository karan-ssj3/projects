```python
##Extracting only the useful feature names that will help in building the recommnedation system
imp_features = ['genres','keywords','tagline','director','cast','title','index']
print(f"Important Feature:{imp_features}")
```

```python
## Checking the Null Values in the dataset
print(f"Genres:{df['genres'].isna().sum()}| Keywords:{df['keywords']}")
```

```python
df = df[imp_features]
```

```python
imp_features = list(df.columns)
imp_features
```

```python
for feature in imp_features:
  df[feature] = df[feature].fillna('')
```

```python
df.isna().sum()
```

```python
combined_features = df['genres'] + df['keywords'] + df['tagline'] + df['director'] + df['cast']
combined_features
```

```python
# converting the text data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

```python
#getting the cosine similarity scores using cosine_similarity
similarity = cosine_similarity(feature_vectors)
```

```python
print(f"Similarity:{similarity},\n Shape:{similarity.shape}")
```

```python
## Taking movie name from use
movie_name = input('Enter your favourtie Movie name:')
```

```python
# creating a list of all movie titles
list_of_all_titles = df['title'].tolist()
print(list_of_all_titles)
```

```python
import difflib
```

```python
#Finding Close match of the movie input
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles )
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
#finding the Index of the movie
index_of_the_movie = df[df.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
df.head()
```

```python
## getting similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
## sorting the movies based on the simialrity score
sorted_movies = sorted(similarity_score, key = lambda x:x[1], reverse =True)
print(sorted_movies)
```

```python
#Printing the name of the similar movies based on index
print("movies suggested for you: \n")
i =1
for movie in sorted_movies:
    index = movie[0]
    title_from_index = df[df.index == index]["title"].values[0]
    if (i<30):
        print(i , ".", title_from_index)
        i +=1
```
